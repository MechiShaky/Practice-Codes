```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_column',None)
```

```python
dataset = pd.read_csv('train.csv')
dataset.head()
```

```python
feature_with_na = [feature for feature in dataset.columns if dataset[feature].isnull().sum() >1]
for feature in feature_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4), '% of missing values')
```

```python
for feature in feature_with_na:
    data= dataset.copy()
    data[feature]= np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
numerical_feature = [feature for feature in dataset.columns if dataset[feature].dtypes !='O']
print('The Number of Numerical Feature ',len(numerical_feature))
```

```python
## Separting Year Feature:
year_feature = [feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
print(year_feature)
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby(feature)['SalePrice'].median().plot()
plt.xlabel('Year')
plt.ylabel('Sale Price')
plt.show()
```

```python
# Relation Between Sale Price Vs Year of Sale
for feature in year_feature:
    if feature != 'YrSold':
        data = dataset.copy()
        data[feature] = data['YrSold'] - data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```

```python
discrete_feature=[feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in year_feature+['Id'] ]
print("The Number of Discreate Variable {}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Sale Price')
    plt.show()
```

```python
continious_feature=[feature for feature in numerical_feature if len(dataset[feature].unique()) > 25 and feature not in year_feature+discrete_feature+['Id']]
print('Length of Continious Feature {}'.format(len(continious_feature)))
```

```python
for feature in continious_feature:
    data= dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

```python
for feature in continious_feature:
    data = dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data['SalePrice'] = np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('Sale Price')
        plt.show()
```
